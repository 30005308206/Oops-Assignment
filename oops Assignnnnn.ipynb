# OOPs Assignment - Theory Answers

## 1. What is Object-Oriented Programming (OOP)?
Object-Oriented Programming (OOP) is a programming paradigm based on the concept of objects that encapsulate data and behavior together.

## 2. What is a class in OOP?
A class is a blueprint for creating objects. It defines attributes (data) and methods (behavior) that the objects created from it will have.

## 3. What is an object in OOP?
An object is an instance of a class. It contains actual data and can perform functions defined in the class.

## 4. Difference between abstraction and encapsulation
- **Abstraction** hides implementation details and shows only essential features.
- **Encapsulation** restricts direct access to some data, ensuring controlled interaction.

## 5. What are dunder methods in Python?
Dunder (double underscore) methods like `__init__`, `__str__`, and `__repr__` are special methods that Python uses to define object behavior.

## 6. Explain the concept of inheritance in OOP
Inheritance allows a class (child) to acquire properties and methods from another class (parent), promoting code reuse.

## 7. What is polymorphism in OOP?
Polymorphism allows methods to take different forms depending on the object calling them, enabling flexibility and reusability.

## 8. How is encapsulation achieved in Python?
Encapsulation is achieved using private (`_var`) and protected (`__var`) attributes, which restrict direct access to class members.

## 9. What is a constructor in Python?
A constructor (`__init__`) is a special method in a class that initializes an object when it is created.

## 10. What are class and static methods in Python?
- **Class methods** (using `@classmethod`) operate on the class itself.
- **Static methods** (using `@staticmethod`) do not operate on the class or instance.

## 11. What is method overloading in Python?
Python does not support method overloading directly but allows default arguments to achieve similar behavior.

## 12. What is method overriding in OOP?
Method overriding allows a child class to redefine a method inherited from the parent class.

## 13. What is a property decorator in Python?
`@property` is a built-in decorator that allows defining getter and setter methods.

## 14. Why is polymorphism important in OOP?
Polymorphism enhances code reusability and flexibility by allowing objects to be treated as instances of their parent class.

## 15. What is an abstract class in Python?
An abstract class cannot be instantiated and must contain at least one abstract method, defined using `@abstractmethod`.

## 16. What are the advantages of OOP?
- Modularity
- Code reusability
- Scalability
- Easier maintenance

## 17. Difference between a class variable and an instance variable
- **Class variables** are shared among all instances.
- **Instance variables** are unique to each object.

## 18. What is multiple inheritance in Python?
Multiple inheritance allows a class to inherit attributes and methods from multiple parent classes.

## 19. Purpose of `__str__` and `__repr__` methods in Python
- `__str__` returns a readable representation.
- `__repr__` returns an unambiguous representation for debugging.

## 20. Significance of the `super()` function in Python
`super()` allows access to parent class methods and attributes in a child class.

## 21. Significance of the `__del__` method in Python
`__del__` is a destructor method called when an object is deleted.

## 22. Difference between `@staticmethod` and `@classmethod`
- `@staticmethod`: Does not require access to class attributes.
- `@classmethod`: Operates on the class itself.

## 23. How does polymorphism work in Python with inheritance?
Child classes can redefine methods inherited from parent classes, allowing different behavior based on the object calling them.

## 24. What is method chaining in Python OOP?
Method chaining allows calling multiple methods on an object in a single statement.

## 25. Purpose of the `__call__` method in Python
The `__call__` method allows an object to be called like a function.
